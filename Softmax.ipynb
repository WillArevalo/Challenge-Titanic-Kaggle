```python
# Softmax o función exponencial normalizada pasa de valores arbitrarios a rangos de 1 a 0
import math
#Tengo cierta lista
z = [1.0, 2.0, 3.0, 4.0, 1.0, 2.0, 3.0]
# Toda mi lista pasa a ser una nueva lista de euler elevado al numero de la lista
z_exp = [math.exp(i) for i in z]
print("e^[z]:",[round(i, 2) for i in z_exp])
# Hago una sumatoria de todos los elementos que han sido elevados
sum_z_exp = sum(z_exp)
print("Sumatoria",round(sum_z_exp, 2))
# Softmax es la lista que se genera a partir de dividir la lista exponencial sobre la sumatoria
softmax = [round(i / sum_z_exp, 3) for i in z_exp]
print("Funcion softmax",softmax)
```

```python
"""Softmax."""

scores = [3.0, 1.0, 0.2]

import numpy as np

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    x_exp = np.exp(x)
    sum_x_exp = np.sum(x_exp, axis=0)
    softmax_result = x_exp / sum_x_exp
    return softmax_result

print(softmax(scores))

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```

```python
"""Softmax."""

scores = [3.0, 1.0, 0.2]

import numpy as np

def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    return np.exp(x) / np.sum(np.exp(x), axis=0)

print(softmax(scores))

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])

plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```
